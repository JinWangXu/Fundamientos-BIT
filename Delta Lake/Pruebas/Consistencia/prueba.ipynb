```
schema = "id INT, nombre STRING, apellido STRING"
# datos de prueba
data = [[1, "Jin", "Wang"], [2, "John","Erik"]]
df = spark.createDataFrame(data, schema)
 # Lo guardamos en una tabla delta
df.write.format("delta").mode('overwrite').save("/dbfs/concurrencia")
df.show()
```

```
# Para realizar un Merge de marcos
schema = "id INT, nombre STRING, apellido STRING"

data = [[3, "Marcos", "Rodriguez"]]
df2 = spark.createDataFrame(data, schema)

df2.write.format("delta").mode('overwrite').save("/dbfs/marcos")
```

```
#Creamos las tablas deltas
df.write.format("delta").mode('overwrite').saveAsTable("concurrencia")
df2.write.format("delta").mode('overwrite').saveAsTable("marcos")
```

```
%sql
SELECT * FROM concurrencia
```

```
%sql
MERGE INTO concurrencia
USING marcos
ON concurrencia.id = marcos.id
WHEN NOT MATCHED
  THEN INSERT *
```

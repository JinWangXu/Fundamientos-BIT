```
df = spark.read.format("delta").load("/dbfs/concurrencia")
df.show()
```

```
# Para merge de marcos 2
schema = "id INT, nombre STRING, apellido STRING"
# El dato es distinto al otro Marcos
data = [[3, "Marcos", "Rodriguez2"]]
df2 = spark.createDataFrame(data, schema)
 # Show the DataFrame; it should reflect our table above
df2.write.format("delta").mode('overwrite').save("/dbfs/marcos2")
```

```
#Creamos las tablas delta
df.write.format("delta").mode('overwrite').saveAsTable("concurrencia")
df2.write.format("delta").mode('overwrite').saveAsTable("marcos2")
```
